```python
# print how many rows are there in the table named 'master'
QUERY = """
SELECT COUNT(*)
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)
df.head()
```

```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)
```

## Demo: How to run a group by query


```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
# your code goes here
# Query to select ConvertedComp from the master table
QUERY = """
SELECT ConvertedComp FROM master
"""

# Retrieve the data
df = pd.read_sql_query(QUERY, conn)

# Plotting the histogram of ConvertedComp
plt.figure(figsize=(10, 6))
plt.hist(df['ConvertedComp'].dropna(), bins=30, edgecolor='black', color='skyblue')

# Add titles and labels
plt.title('Histogram of Converted Compensation', fontsize=16)
plt.xlabel('Converted Compensation (USD)', fontsize=12)
plt.ylabel('Frequency', fontsize=12)

# Show the plot
plt.show()

```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here

# Query to select Age from the master table
QUERY = """
SELECT Age FROM master
"""

# Retrieve the data
df = pd.read_sql_query(QUERY, conn)

# Plotting the box plot of Age
plt.figure(figsize=(10, 6))
plt.boxplot(df['Age'].dropna(), patch_artist=True, boxprops=dict(facecolor='lightblue', color='black'))

# Add titles and labels
plt.title('Box Plot of Age', fontsize=16)
plt.ylabel('Age', fontsize=12)

# Show the plot
plt.show()
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here
# Query to select Age and WorkWeekHrs from the master table
QUERY = """
SELECT Age, WorkWeekHrs FROM master
"""

# Retrieve the data
df = pd.read_sql_query(QUERY, conn)

# Plotting the scatter plot of Age vs WorkWeekHrs
plt.figure(figsize=(10, 6))
plt.scatter(df['Age'], df['WorkWeekHrs'], alpha=0.5, color='blue')

# Add titles and labels
plt.title('Scatter Plot of Age vs WorkWeekHrs', fontsize=16)
plt.xlabel('Age', fontsize=12)
plt.ylabel('Work Week Hours', fontsize=12)

# Show the plot
plt.grid()
plt.show()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here
# Query to select WorkWeekHrs, CodeRevHrs, and Age from the master table
QUERY = """
SELECT WorkWeekHrs, CodeRevHrs, Age FROM master
"""

# Retrieve the data
df = pd.read_sql_query(QUERY, conn)

# Check for NaN values and remove them for plotting
df = df.dropna(subset=['WorkWeekHrs', 'CodeRevHrs', 'Age'])

# Create a bubble plot
plt.figure(figsize=(10, 6))
plt.scatter(df['WorkWeekHrs'], df['CodeRevHrs'], s=df['Age']*10, alpha=0.5, color='green', edgecolors='w', linewidth=0.5)

# Add titles and labels
plt.title('Bubble Plot of WorkWeekHrs vs CodeRevHrs', fontsize=16)
plt.xlabel('Work Week Hours', fontsize=12)
plt.ylabel('Code Review Hours', fontsize=12)

# Show the plot
plt.grid()
plt.show()

```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# your code goes here
# Query to get the top 5 databases respondents wish to learn
QUERY = """
SELECT DatabaseDesireNextYear, COUNT(*) AS count
FROM DatabaseDesireNextYear
WHERE DatabaseDesireNextYear IS NOT NULL
GROUP BY DatabaseDesireNextYear
ORDER BY count DESC
LIMIT 5;
"""

# Execute the query and read the result into a DataFrame
top_databases = pd.read_sql_query(QUERY, conn)

# Plotting the pie chart
plt.figure(figsize=(8, 8))
plt.pie(top_databases['count'], labels=top_databases['DatabaseDesireNextYear'], autopct='%1.1f%%', startangle=140)
plt.title('Top 5 Databases Respondents Wish to Learn Next Year', fontsize=16)
plt.axis('equal')

# Show the pie chart
plt.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here

# Query to select Age, WorkWeekHrs, and CodeRevHrs from the master table
QUERY = """
SELECT Age, WorkWeekHrs, CodeRevHrs FROM master
WHERE Age BETWEEN 30 AND 35
"""

# Retrieve the data
df = pd.read_sql_query(QUERY, conn)

# Calculate the median for WorkWeekHrs and CodeRevHrs
median_values = df[['WorkWeekHrs', 'CodeRevHrs']].median()

# Create a stacked bar chart
plt.figure(figsize=(10, 6))
plt.bar(median_values.index, median_values.values, color=['skyblue', 'salmon'])

# Add titles and labels
plt.title('Median WorkWeekHrs and CodeRevHrs for Age Group 30 to 35', fontsize=16)
plt.xlabel('Hour Categories', fontsize=12)
plt.ylabel('Median Hours', fontsize=12)

# Add the median values on top of the bars
for index, value in enumerate(median_values):
    plt.text(index, value, f'{value:.1f}', ha='center', va='bottom')

# Show the plot
plt.tight_layout()
plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here
# Query to select Age and ConvertedComp from the master table
QUERY = """
SELECT Age, ConvertedComp FROM master
WHERE Age BETWEEN 45 AND 60
"""

# Retrieve the data
df = pd.read_sql_query(QUERY, conn)

# Calculate the median ConvertedComp for each age
median_converted_comp = df.groupby('Age')['ConvertedComp'].median().reset_index()

# Create a bar chart for the median ConvertedComp by Age
plt.figure(figsize=(10, 6))
plt.bar(median_converted_comp['Age'], median_converted_comp['ConvertedComp'], color='lightgreen')

# Add titles and labels
plt.title('Median ConvertedComp for Ages 45 to 60', fontsize=16)
plt.xlabel('Age', fontsize=12)
plt.ylabel('Median Converted Compensation (USD)', fontsize=12)

# Add the median values on top of the bars
for index, value in enumerate(median_converted_comp['ConvertedComp']):
    plt.text(median_converted_comp['Age'][index], value, f'{value:.2f}', ha='center', va='bottom')

# Show the plot
plt.tight_layout()
plt.show()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here
# Query to select the MainBranch column and count occurrences
QUERY = """
SELECT MainBranch, COUNT(*) as count
FROM master
GROUP BY MainBranch
"""

# Retrieve the data
df_mainbranch = pd.read_sql_query(QUERY, conn)

# Create a horizontal bar chart for MainBranch
plt.figure(figsize=(10, 6))
plt.barh(df_mainbranch['MainBranch'], df_mainbranch['count'], color='skyblue')

# Add titles and labels
plt.title('Count of Respondents by Main Branch', fontsize=16)
plt.xlabel('Count of Respondents', fontsize=12)
plt.ylabel('Main Branch', fontsize=12)

# Show the plot
plt.tight_layout()
plt.show()
```
